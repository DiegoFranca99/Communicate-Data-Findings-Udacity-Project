## Dataset Overview

> A dataset about video games sales extract from kaggle with more than 16500 games. The dataset have collumns with name, platform, year, genre, publisher, north america sale, europe sales, japan sales, other region sales and global sales, all sales information are in milions dollars scale. 

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
games = pd.read_csv('vgsales.csv')
default_color = sb.color_palette()[0] #To set a default color to all plots.
````

## Step 1

> Which genre are the most popular all data set? Should the most popular be the best seller?

````python
freq = games['Genre'].value_counts()
gnr_order = freq.index
plt.figure(figsize=[16, 9])
sb.countplot(data = games, x = 'Genre', color = default_color, order = gnr_order)
plt.xlabel('Genre')
plt.title('Count of video games per genre');
plt.show()
````

## Step 2

> As we saw, Action is the most popular video game genre. Now that we know that, let's add Global Sales variable to find out which one are the best seller and if it is action or not.

````python
base_color = sb.color_palette()[0]
plt.figure(figsize=[16, 9])
sb.boxplot(data=games, x='Genre', y='Global_Sales', color=base_color);
plt.title('Box plot of global sales per genre');
````

## Step 3

>Alright, it looked that action got down to the 9th position loosing to sports on a global scale. Now, let's plot adding publisher variable to see which one is the best selles as well and to go a litlle depper, let's break in a region wise? Is it different than before? To better visualizarion, let's plot only the Global sales greather than 10 milion.

````python
outliers_global_genre = (games['Global_Sales'] > 10)
plt.figure(figsize=[16, 9])
plt.subplot(3, 1, 1)
ax = sb.barplot(data = games.loc[outliers_global_genre,:], x = 'Publisher', y = 'NA_Sales', hue = 'Genre')
ax.legend(loc = 1, ncol = 3, framealpha = 1, title = 'Genre');
plt.title('Bar chart of global sales per genre per publisher and region of the 62 best seller video games ');
plt.subplot(3, 1, 2)
ax = sb.barplot(data = games.loc[outliers_global_genre,:], x = 'Publisher', y = 'EU_Sales', hue = 'Genre')
ax.legend(loc = 1, ncol = 3, framealpha = 1, title = 'Genre');
plt.subplot(3, 1, 3)
ax = sb.barplot(data = games.loc[outliers_global_genre,:], x = 'Publisher', y = 'JP_Sales', hue = 'Genre')
ax.legend(loc = 1, ncol = 3, framealpha = 1, title = 'Genre');
````

## Conclusion

### From the previous screen, it is possible to describe the results. Nintendo turned out to be the greather publisher on all principal region. But when it comes to genre, Noth America and Europe region share the same most popular genre: Sports, but Japan region stick with Role-Playing insted. A interest result since the most popular genre on data set are Action, that does not appear on the most on this perspective.
